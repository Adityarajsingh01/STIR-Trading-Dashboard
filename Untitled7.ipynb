```python
# @title 🏛️ STIR Pro Dashboard: Trader's Edition (Final Refined)
# @markdown ### **Trader's Guide**
# @markdown 1. **Visuals**: Excel Export now includes **Native Charts** automatically.
# @markdown 2. **Strategies**: "Deep Dive" tab now auto-calculates **Flys & Spreads**.
# @markdown 3. **Logic**: Exact T+1 settlement with Basis (ZQ-SR1) monitoring.

import subprocess
import sys

# --- AUTO-INSTALL DEPENDENCY ---
def install(package):
    try:
        import xlsxwriter
    except ImportError:
        print(f"⚙️ Installing system dependency: {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

install('xlsxwriter')
import xlsxwriter

# --- MAIN DASHBOARD CODE ---
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
import ipywidgets as widgets
from IPython.display import display, clear_output
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
from datetime import date, timedelta
import io

# Professional Dark Theme
pio.templates.default = "plotly_dark"
pd.options.display.float_format = '{:,.3f}'.format

# ==========================================
# 1. CORE LOGIC ENGINE
# ==========================================

class MarketDataEngine:
    def __init__(self):
        # FOMC Decision Dates
        self.fomc_schedule = {
            2024: ['2024-01-31', '2024-03-20', '2024-05-01', '2024-06-12', '2024-07-31', '2024-09-18', '2024-11-07', '2024-12-18'],
            2025: ['2025-01-29', '2025-03-19', '2025-05-07', '2025-06-18', '2025-07-30', '2025-09-17', '2025-10-29', '2025-12-10'],
            2026: ['2026-01-28', '2026-03-18', '2026-04-29', '2026-06-17', '2026-07-29', '2026-09-16', '2026-10-28', '2026-12-09'],
            2027: ['2027-01-27', '2027-03-17', '2027-04-28', '2027-06-09', '2027-07-28', '2027-09-15', '2027-10-27', '2027-12-08']
        }
        self.cal = USFederalHolidayCalendar()
        self.bd = CustomBusinessDay(calendar=self.cal)

    def get_effective_date(self, meeting_date_str):
        """Returns T+1 Business Day (Effective Date)"""
        return pd.to_datetime(meeting_date_str) + self.bd

    def generate_impact_matrix(self, year):
        """Calculates Impact Factors (Sensitivity)"""
        start = pd.Timestamp(f"{year}-01-01")
        end = pd.Timestamp(f"{year}-12-31")

        meetings = self.fomc_schedule.get(year, [])
        matrix_data = []

        for m_str in meetings:
            eff_date = self.get_effective_date(m_str)
            row = {'Meeting': m_str, 'Effective': eff_date.strftime('%Y-%m-%d')}

            for month_idx in range(1, 13):
                m_start = pd.Timestamp(f"{year}-{month_idx:02d}-01")
                m_end = m_start + pd.offsets.MonthEnd(0)
                total_days = (m_end - m_start).days + 1

                overlap_start = max(eff_date, m_start)
                overlap_end = min(end, m_end)

                if overlap_start <= overlap_end:
                    days_affected = (overlap_end - overlap_start).days + 1
                    factor = days_affected / total_days
                else:
                    factor = 0.0

                col_name = m_start.strftime('%b')
                row[col_name] = factor
            matrix_data.append(row)

        return pd.DataFrame(matrix_data)

    def generate_daily_curve(self, year, effr_start, sofr_start, hikes_map, turn_premiums):
        """Generates Daily Rates (Staircase)"""
        start_date = pd.Timestamp(f"{year}-01-01")
        end_date = pd.Timestamp(f"{year}-12-31")
        dates = pd.date_range(start=start_date, end=end_date, freq='D')

        effective_changes = {}
        for m_date, change in hikes_map.items():
            eff_date = self.get_effective_date(m_date)
            effective_changes[eff_date] = change

        sorted_changes = sorted(effective_changes.keys())
        change_series = pd.Series(0.0, index=dates)

        for d in sorted_changes:
            if start_date <= d <= end_date:
                change_series.loc[d] = effective_changes[d]

        rate_path_effr = change_series.cumsum() + effr_start
        rate_path_sofr = change_series.cumsum() + sofr_start

        final_effr = rate_path_effr.copy()
        final_sofr = rate_path_sofr.copy()

        for d in dates:
            addon = 0.0
            if d.is_year_end:
                addon += turn_premiums.get('Year End', 0)
            elif d.is_quarter_end:
                addon += turn_premiums.get('Quarter End', 0)
            elif d.is_month_end:
                addon += turn_premiums.get('Month End', 0)

            if addon != 0:
                final_effr.loc[d] += addon
                final_sofr.loc[d] += addon

        return pd.DataFrame({
            'Date': dates,
            'EFFR': final_effr.values,
            'SOFR': final_sofr.values
        })

    def calculate_pricing(self, daily_df):
        """Aggregates Daily -> Monthly"""
        daily_df['Month_ID'] = daily_df['Date'].dt.to_period('M')

        monthly = daily_df.groupby('Month_ID').agg(
            Avg_EFFR=('EFFR', 'mean'),
            Avg_SOFR=('SOFR', 'mean')
        ).reset_index()

        monthly['ZQ'] = 100 - monthly['Avg_EFFR']
        monthly['SR1'] = 100 - monthly['Avg_SOFR']

        # Basis Calculation (ZQ - SR1) * 100 for bps
        monthly['Basis_bps'] = (monthly['ZQ'] - monthly['SR1']) * 100

        monthly['Month_Label'] = monthly['Month_ID'].dt.strftime('%b %Y')

        codes = {1:'F', 2:'G', 3:'H', 4:'J', 5:'K', 6:'M', 7:'N', 8:'Q', 9:'U', 10:'V', 11:'X', 12:'Z'}
        monthly['Code'] = monthly['Month_ID'].dt.month.map(codes)

        return monthly

    def calculate_strategies(self, pricing):
        # 1. Calendar Spreads
        spreads = []
        for i in range(len(pricing)-1):
            s = pricing.iloc[i]
            e = pricing.iloc[i+1]
            val = (s['SR1'] - e['SR1']) * 100
            spreads.append({'Strategy': f"{s['Code']}/{e['Code']}", 'Price': val, 'Type': 'Spread'})

        # 2. Butterflys (Quarterly)
        q_months = [3, 6, 9, 12]
        q_df = pricing[pricing['Month_ID'].dt.month.isin(q_months)].reset_index(drop=True)
        for i in range(len(q_df)-2):
            w1, body, w2 = q_df.iloc[i], q_df.iloc[i+1], q_df.iloc[i+2]
            val = (2*body['SR1'] - (w1['SR1'] + w2['SR1'])) * 100
            spreads.append({'Strategy': f"{body['Code']} Fly", 'Price': val, 'Type': 'Fly'})

        return pd.DataFrame(spreads)

# ==========================================
# 2. SESSION & EXCEL ENGINE (WITH CHARTS)
# ==========================================

class SessionManager:
    def __init__(self):
        self.scenarios = {}
        self.active_data = None

    def save_scenario(self, name, pricing, daily, impact, params):
        self.scenarios[name] = {
            'pricing': pricing.copy(),
            'daily': daily.copy(),
            'impact': impact.copy(),
            'params': params
        }

    def generate_master_excel(self):
        filename = "STIR_Master_Analysis.xlsx"

        with pd.ExcelWriter(filename, engine='xlsxwriter') as writer:
            workbook = writer.book

            # Formats
            fmt_pct = workbook.add_format({'num_format': '0.000%', 'align': 'center'})
            fmt_px = workbook.add_format({'num_format': '0.000', 'align': 'center'})
            fmt_bps = workbook.add_format({'num_format': '+0.0', 'align': 'center'})

            # 1. SCENARIO COMPARISON SHEET
            if self.scenarios:
                comp_df = None
                for name, data in self.scenarios.items():
                    df = data['pricing'][['Month_Label', 'SR1']].copy()
                    df.columns = ['Month', f'{name} (SR1)']
                    if comp_df is None:
                        comp_df = df
                    else:
                        comp_df = pd.merge(comp_df, df, on='Month')

                # Delta
                cols = [c for c in comp_df.columns if c != 'Month']
                if len(cols) >= 2:
                    delta_col = f'Diff ({cols[1]} - {cols[0]}) bps'
                    comp_df[delta_col] = (comp_df[cols[1]] - comp_df[cols[0]]) * 100

                comp_df.to_excel(writer, sheet_name='Comparison', index=False)
                ws = writer.sheets['Comparison']
                ws.set_column(0, len(comp_df.columns), 15, fmt_px)

                # --- CHART: Scenario Comparison ---
                chart = workbook.add_chart({'type': 'line'})
                max_row = len(comp_df) + 1
                # Add series for each scenario (skipping Month and Delta columns)
                for i, col_name in enumerate(cols):
                    chart.add_series({
                        'name':       col_name,
                        'categories': ['Comparison', 1, 0, max_row-1, 0],
                        'values':     ['Comparison', 1, i+1, max_row-1, i+1],
                        'line':       {'width': 2.25},
                    })
                chart.set_title({'name': 'Scenario Pricing Comparison'})
                chart.set_y_axis({'name': 'Price'})
                ws.insert_chart('F2', chart)

            # 2. ACTIVE SCENARIO SHEETS
            if self.active_data:
                p_df = self.active_data['pricing']
                d_df = self.active_data['daily']

                export_cols = ['Month_Label', 'Code', 'ZQ', 'SR1', 'Basis_bps', 'Avg_EFFR', 'Avg_SOFR']
                p_df[export_cols].to_excel(writer, sheet_name='Active_Pricing', index=False)

                ws_act = writer.sheets['Active_Pricing']
                ws_act.set_column('A:G', 12, fmt_px)

                # --- CHART: Active Curve ---
                chart_act = workbook.add_chart({'type': 'line'})
                max_row = len(p_df) + 1
                chart_act.add_series({
                    'name':       'SR1 Price',
                    'categories': ['Active_Pricing', 1, 0, max_row-1, 0], # Month Label
                    'values':     ['Active_Pricing', 1, 3, max_row-1, 3], # SR1 Col
                    'line':       {'color': '#00B050', 'width': 2.25},
                })
                chart_act.set_title({'name': 'Current Pricing Curve'})
                ws_act.insert_chart('I2', chart_act)

                d_df.to_excel(writer, sheet_name='Active_Daily_Rates', index=False)

        return filename

session = SessionManager()
engine = MarketDataEngine()

# ==========================================
# 3. DASHBOARD UI
# ==========================================

style_header = "<div style='background-color:#0D1117; color:#58A6FF; padding:15px; border-radius:10px; border:1px solid #30363D; text-align:center'><h2>⚡ STIR TRADER PRO <span style='font-size:14px; color:#8B949E'>// ANALYTICS SUITE</span></h2></div>"
style_section = "<div style='color:#58A6FF; font-weight:bold; margin-top:15px; margin-bottom:10px; border-bottom:1px solid #30363D; padding-bottom:5px;'>{}</div>"

w_year = widgets.Dropdown(options=[2024, 2025, 2026, 2027], value=2026, description='Year:')
w_effr = widgets.FloatText(value=5.33, description='Start EFFR:', step=0.01)
w_sofr = widgets.FloatText(value=5.33, description='Start SOFR:', step=0.01)

w_tp_m = widgets.FloatText(value=0.0, description='M-End (bps):')
w_tp_q = widgets.FloatText(value=10.0, description='Q-End (bps):')
w_tp_y = widgets.FloatText(value=25.0, description='Y-End (bps):')

w_scenario_name = widgets.Text(value='Base Case', description='Name:')
hikes_container = widgets.VBox([])

def refresh_hikes_ui(change=None):
    year = w_year.value
    dates = engine.fomc_schedule.get(year, [])
    items = [widgets.HTML(f"<br><b>FOMC Meetings {year} (Input bps +/-)</b>")]
    grid = []
    for d in dates:
        d_lbl = pd.to_datetime(d).strftime('%b %d')
        ft = widgets.FloatText(value=0.0, description=d_lbl, layout=widgets.Layout(width='140px'), style={'description_width': '50px'})
        ft.meeting_date = d
        grid.append(ft)
    rows = [widgets.HBox(grid[i:i+4]) for i in range(0, len(grid), 4)]
    items.extend(rows)
    hikes_container.children = tuple(items)

w_year.observe(refresh_hikes_ui, names='value')

# Strategy Monitor Widgets
w_spread_type = widgets.ToggleButtons(options=['SR1', 'ZQ'], value='SR1', description='Product:')
out_chart = widgets.Output()
out_table = widgets.Output()
out_spreads = widgets.Output() # Now handles manual + auto
out_matrix = widgets.Output()
out_compare_chart = widgets.Output()
out_compare_table = widgets.Output()
out_export_status = widgets.Output()

def run_calculation():
    hikes = {}
    for row in hikes_container.children[1:]:
        if isinstance(row, widgets.HBox):
            for w in row.children:
                hikes[w.meeting_date] = w.value / 100.0

    turns = {'Month End': w_tp_m.value/100, 'Quarter End': w_tp_q.value/100, 'Year End': w_tp_y.value/100}

    daily = engine.generate_daily_curve(w_year.value, w_effr.value, w_sofr.value, hikes, turns)
    pricing = engine.calculate_pricing(daily)
    impact = engine.generate_impact_matrix(w_year.value)
    strategies = engine.calculate_strategies(pricing)

    session.active_data = {
        'pricing': pricing,
        'daily': daily,
        'impact': impact,
        'params': {'year': w_year.value, 'hikes': hikes},
        'strategies': strategies
    }
    return pricing, daily, impact, strategies

def on_run_click(b):
    out_chart.clear_output()
    out_table.clear_output()
    out_spreads.clear_output()
    out_matrix.clear_output()

    pricing, daily, impact, strategies = run_calculation()

    with out_chart:
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=daily['Date'], y=daily['EFFR'], name='EFFR', line_shape='hv', line=dict(color='#00E5FF')))
        fig.add_trace(go.Scatter(x=daily['Date'], y=daily['SOFR'], name='SOFR', line_shape='hv', line=dict(color='#E040FB', dash='dot')))
        fig.update_layout(title="Projected Rates (Step Accrual)", height=350, margin=dict(l=20, r=20, t=30, b=20))
        fig.show()

    with out_table:
        cols = ['Month_Label', 'ZQ', 'SR1', 'Basis_bps', 'Avg_EFFR']
        styler = pricing[cols].style.format({
            'ZQ':'{:.3f}',
            'SR1':'{:.3f}',
            'Basis_bps':'{:+.1f}',
            'Avg_EFFR':'{:.3f}%'
        }).background_gradient(cmap='RdYlGn', subset=['Basis_bps'])
        display(styler)

    with out_spreads:
        # Auto Strategy Ladder
        display(widgets.HTML("<b>Strategy Ladder (Spreads & Flys)</b>"))
        st_style = strategies.style.format({'Price': '{:+.1f}'}).bar(subset=['Price'], align='mid', color=['#FF5252', '#69F0AE'])
        display(st_style)

    with out_matrix:
        display(widgets.HTML("<b>Impact Matrix (Sensitivity)</b>"))
        display(impact.style.format("{:.1%}", subset=impact.columns[2:]).background_gradient(cmap='Blues'))

def update_compare_tab():
    out_compare_chart.clear_output()
    out_compare_table.clear_output()

    if not session.scenarios: return

    with out_compare_chart:
        fig = go.Figure()
        for name, data in session.scenarios.items():
            df = data['pricing']
            fig.add_trace(go.Scatter(x=df['Month_Label'], y=df['SR1'], mode='lines+markers', name=name))
        fig.update_layout(title="Scenario Comparison (SR1 Pricing)", yaxis_title="Price", height=450)
        fig.show()

    with out_compare_table:
        comp_df = None
        for name, data in session.scenarios.items():
            s = data['pricing'][['Month_Label', 'SR1']].copy()
            s.rename(columns={'SR1': name}, inplace=True)
            if comp_df is None: comp_df = s
            else: comp_df = pd.merge(comp_df, s, on='Month_Label')

        cols = [c for c in comp_df.columns if c != 'Month_Label']
        if len(cols) >= 2:
            delta_col = f'Diff ({cols[1]} - {cols[0]}) bps'
            comp_df[delta_col] = (comp_df[cols[1]] - comp_df[cols[0]]) * 100

            styler = comp_df.style.format("{:.3f}", subset=cols).format("{:+.1f}", subset=[delta_col])
            styler.bar(subset=[delta_col], align='mid', color=['#FF5252', '#69F0AE'])
            display(styler)
        else:
            display(comp_df.style.format("{:.3f}", subset=cols))

def on_save_click(b):
    if not session.active_data:
        on_run_click(None)

    name = w_scenario_name.value
    d = session.active_data
    session.save_scenario(name, d['pricing'], d['daily'], d['impact'], d['params'])

    update_compare_tab()

    with out_chart:
        print(f"✅ Scenario '{name}' Saved! Check 'Comparison' tab.")

def on_export_master(b):
    out_export_status.clear_output()
    if not session.scenarios and not session.active_data:
        with out_export_status: print("❌ Run analysis or save a scenario first.")
        return

    fname = session.generate_master_excel()

    with out_export_status:
        try:
            from google.colab import files
            files.download(fname)
            display(widgets.HTML(f"<div style='color:#58A6FF; border:1px solid #30363D; padding:10px;'>✅ <b>{fname}</b> ready! Check downloads.</div>"))
        except:
            print(f"Saved {fname} locally.")

btn_run = widgets.Button(description='Run Analysis', button_style='primary', icon='play')
btn_save = widgets.Button(description='💾 Save to Session', button_style='warning')
btn_export = widgets.Button(description='📥 Export Master Workbook', button_style='success', icon='table', layout=widgets.Layout(width='100%'))

btn_run.on_click(on_run_click)
btn_save.on_click(on_save_click)
btn_export.on_click(on_export_master)

t1 = widgets.VBox([
    widgets.HTML(style_section.format("1. Inputs")),
    widgets.HBox([widgets.VBox([w_year, w_effr, w_sofr]), widgets.VBox([w_tp_m, w_tp_q, w_tp_y])]),
    hikes_container,
    widgets.HTML("<hr>"),
    widgets.HBox([btn_run, w_scenario_name, btn_save]),
    out_chart,
    out_table
])

t2 = widgets.VBox([
    widgets.HTML(style_section.format("2. Strategy Ladder")),
    widgets.HTML("<i>Displays calculated Calendar Spreads & Quarterly Flys</i>"),
    out_spreads,
    widgets.HTML(style_section.format("3. Sensitivity Matrix")),
    out_matrix
])

t3 = widgets.VBox([
    widgets.HTML(style_section.format("4. Strategy Comparison")),
    btn_export,
    out_export_status,
    out_compare_chart,
    out_compare_table
])

tabs = widgets.Tab([t1, t2, t3])
tabs.set_title(0, 'Builder')
tabs.set_title(1, 'Deep Dive')
tabs.set_title(2, 'Comparison & Export')

refresh_hikes_ui()
display(widgets.HTML(style_header))
display(tabs)
```
